A 60 g mass with a horizontal velocity of 50 cm/s collides with an 85 g mass at rest. After collision, the 60 g mass travel 30° below the horizontal and the 85 g mass travels 45° above the horizontal. Find the speed of each mass after impact if collision is perfectly elastic.


```python
import matplotlib.pyplot as plt
import numpy as np

# Set up the figure with a dark background
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(10, 8))

# Initial positions
ax.plot(0, 0, 'o', ms=10, c='cyan', label='60 g (50 cm/s)')
ax.plot(2, 0, 'o', ms=12, c='yellow', label='85 g (rest)')

# Initial velocity
ax.arrow(0, 0, 0.5, 0, head_width=0.1, head_length=0.1, fc='cyan', ec='cyan')
ax.text(0.6, 0.1, '50 cm/s', color='cyan', fontsize=12)

# Collision point
ax.plot(2, 0, 'x', ms=10, c='white', label='Collision')

# Post-collision paths
# Mass 1: 30° below
theta1 = np.deg2rad(-30)
ax.arrow(2, 0, 0.366 * np.cos(theta1), 0.366 * np.sin(theta1),
         head_width=0.1, head_length=0.1, fc='cyan', ec='cyan')
ax.text(2.4, -0.3, '36.6 cm/s\n(-30°)', color='cyan', fontsize=12)

# Mass 2: 45° above
theta2 = np.deg2rad(45)
ax.arrow(2, 0, 0.183 * np.cos(theta2), 0.183 * np.sin(theta2),
         head_width=0.1, head_length=0.1, fc='yellow', ec='yellow')
ax.text(2.3, 0.3, '18.3 cm/s\n(45°)', color='yellow', fontsize=12)

# Styling
ax.set_xlim(-1, 3)
ax.set_ylim(-1, 1)
ax.set_aspect('equal')
ax.set_xlabel('x (m)', fontsize=14, color='white')
ax.set_ylabel('y (m)', fontsize=14, color='white')
ax.set_title('Perfectly Elastic Collision', fontsize=16, color='white')
ax.legend(fontsize=12, loc='upper left')
ax.grid(True, ls='--', alpha=0.3)

# Remove spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('white')
ax.spines['bottom'].set_color('white')

plt.show()
```

A 3 kg block moving to the right on a frictionless table at 4 m/s makes a head on collision with another 3 kg block moving 5 m/s to the left.

a) If the collision is completely inelastic, find the final velocity of the blocks.

b) If the collision is completely elastic, find the final velocity of the blocks.

c) If half the initial energy is dissipated in the collision, find the final velocities of the
blocks.

```python
# Set up the figure with a dark background
plt.style.use('dark_background')
fig, ax = plt.subplots(3, 1, figsize=(10, 12), sharex=True)

# Common settings
titles = ['(a) Completely Inelastic', '(b) Completely Elastic', '(c) 50% Energy Dissipated']
initial_pos = [(0, 0), (2, 0)]  # Block 1, Block 2
colors = ['cyan', 'yellow']
labels = ['3 kg (4 m/s)', '3 kg (-5 m/s)']
final_velocities = [
    [(-0.5, -0.5)],  # Inelastic: both at -0.5 m/s
    [(-5, 4)],       # Elastic: -5, 4 m/s
    [(-3.66, 2.66)]  # Partial: -3.66, 2.66 m/s
]

for i in range(3):
    # Table
    ax[i].plot([-2, 4], [0, 0], 'w-', lw=2, label='Frictionless Table')

    # Initial blocks
    ax[i].plot(initial_pos[0][0], 0, 's', ms=15, c=colors[0], label=labels[0])
    ax[i].plot(initial_pos[1][0], 0, 's', ms=15, c=colors[1], label=labels[1])

    # Initial velocities
    ax[i].arrow(0, 0, 0.4, 0, head_width=0.05, head_length=0.1, fc=colors[0], ec=colors[0])
    ax[i].arrow(2, 0, -0.5, 0, head_width=0.05, head_length=0.1, fc=colors[1], ec=colors[1])

    # Collision point
    ax[i].plot(1, 0, 'x', ms=10, c='white', label='Collision')

    # Final velocities
    if i == 0:  # Inelastic
        ax[i].arrow(1, 0, -0.2, 0, head_width=0.05, head_length=0.1, fc='gray', ec='gray')
        ax[i].text(1.5, 0.2, 'U = -0.5 m/s', color='gray', fontsize=12)
    else:
        # Block 1
        scale1 = 0.3 if final_velocities[i][0][0] < 0 else -0.3
        ax[i].arrow(1, 0, scale1, 0, head_width=0.05, head_length=0.1, fc=colors[0], ec=colors[0])
        ax[i].text(1.5 if scale1 < 0 else 0.5, 0.2, f'U₁ = {final_velocities[i][0][0]:.2f} m/s',
                   color=colors[0], fontsize=12)
        # Block 2
        scale2 = 0.3 if final_velocities[i][0][1] > 0 else -0.3
        ax[i].arrow(1, 0, scale2, 0, head_width=0.05, head_length=0.1, fc=colors[1], ec=colors[1])
        ax[i].text(1.5 if scale2 < 0 else 0.5, -0.2, f'U₂ = {final_velocities[i][0][1]:.2f} m/s',
                   color=colors[1], fontsize=12)

    # Styling
    ax[i].set_xlim(-2, 4)
    ax[i].set_ylim(-0.5, 0.5)
    ax[i].set_aspect('equal')
    ax[i].set_yticks([])
    ax[i].set_title(titles[i], fontsize=14, color='white')
    ax[i].legend(fontsize=10, loc='upper left')
    ax[i].grid(True, ls='--', alpha=0.3)

    # Remove spines
    ax[i].spines['top'].set_visible(False)
    ax[i].spines['right'].set_visible(False)
    ax[i].spines['left'].set_visible(False)
    ax[i].spines['bottom'].set_color('white')

ax[2].set_xlabel('x (m)', fontsize=14, color='white')
plt.tight_layout()
plt.show()
```

a) What is the momentum of a 12000 kg truck whose velocity is 50 m/s?

b) What velocity must a 6000 kg truck attain in order to have the same momentum?

c) What velocity must it attain to have the same kinetic energy?

```python
# Set up the figure with a dark background
import matplotlib.pyplot as plt
import numpy as np

# Set 3B1B-inspired style
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(12, 8))

# Road
ax.plot([-2, 10], [0, 0], 'w-', lw=3, label='Road')

# Truck 1
truck1 = plt.Rectangle((0, 0), 2, 1, fc='cyan', ec='white', label='Truck 1 (12,000 kg)')
ax.add_patch(truck1)
ax.arrow(2, 0.5, 1, 0, head_width=0.1, head_length=0.2, fc='cyan', ec='cyan')
ax.text(3, 0.7, 'v₁ = 50 m/s\np₁ = 600,000 kg·m/s\nKE₁ = 15,000,000 J',
        color='cyan', fontsize=12)

# Truck 2 (Momentum)
truck2_mom = plt.Rectangle((4, 1.5), 2, 1, fc='yellow', ec='white', label='Truck 2 (6,000 kg, Same p)')
ax.add_patch(truck2_mom)
ax.arrow(6, 2, 1, 0, head_width=0.1, head_length=0.2, fc='yellow', ec='yellow')
ax.text(7, 2.2, 'v₂ = 100 m/s\np₂ = 600,000 kg·m/s',
        color='yellow', fontsize=12)

# Truck 2 (Kinetic Energy)
truck2_ke = plt.Rectangle((4, -1.5), 2, 1, fc='red', ec='white', label='Truck 2 (6,000 kg, Same KE)')
ax.add_patch(truck2_ke)
ax.arrow(6, -1, 1, 0, head_width=0.1, head_length=0.2, fc='red', ec='red')
ax.text(7, -0.8, 'v₂ = 70.71 m/s\nKE₂ = 15,000,000 J',
        color='red', fontsize=12)

# Styling
ax.set_xlim(-2, 10)
ax.set_ylim(-2.5, 3.5)
ax.set_aspect('equal')
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel('Direction of Motion', fontsize=14, color='white')
ax.set_ylabel('Vertical', fontsize=14, color='white')
ax.set_title('Truck Momentum and Kinetic Energy Comparison', fontsize=16, color='white')
ax.legend(fontsize=12, loc='upper left')
ax.grid(True, ls='--', alpha=0.3)

# Remove spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('white')

plt.show()
```

A bullet, having a mass of 0.04 kg moving with a velocity of 450 m/s penetrates a distance of 0.08 m into a wooden block firmly attached to the earth. Assuming a constant accelerating force, compute the following: a) the acceleration of the bullet, b) the accelerating force, c) the time of acceleration, and d) the impulse of the collision.

```python
# Set up the figure with a dark background
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(12, 6))

# Block
block = plt.Rectangle((2, -0.5), 4, 2, fc='gray', ec='white', label='Wooden Block (Fixed)')
ax.add_patch(block)

# Bullet (initial)
ax.plot(0, 0.5, 'o', ms=10, c='cyan', label='Bullet (0.04 kg)')
ax.arrow(0, 0.5, 1, 0, head_width=0.1, head_length=0.2, fc='cyan', ec='cyan')
ax.text(1.5, 0.7, 'v = 450 m/s', color='cyan', fontsize=12)

# Bullet (final)
ax.plot(2.08, 0.5, 'o', ms=10, c='red')
ax.text(2.3, 0.7, 'v = 0\ns = 0.08 m', color='red', fontsize=12)

# Force
ax.arrow(2.08, 0.5, -0.5, 0, head_width=0.1, head_length=0.2, fc='yellow', ec='yellow')
ax.text(1.5, 0.3, 'F = 50,625 N', color='yellow', fontsize=12)

# Results
ax.text(4, 1.5, 'a = -1.266×10⁶ m/s²\nt = 3.556×10⁻⁴ s\nJ = 18 N·s',
        color='white', fontsize=12, ha='center',
        bbox=dict(facecolor='black', alpha=0.5))

# Styling
ax.set_xlim(-1, 7)
ax.set_ylim(-1, 2)
ax.set_aspect('equal')
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel('Direction of Motion', fontsize=14, color='white')
ax.set_ylabel('Vertical', fontsize=14, color='white')
ax.set_title('Bullet Penetrating Wooden Block', fontsize=16, color='white')
ax.legend(fontsize=12, loc='upper left')
ax.grid(True, ls='--', alpha=0.3)

# Remove spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('white')

plt.show()
```

A steel ball of mass 0.514 kg is fastened to a cord 68.7 cm long and is released when the cord is horizontal. At the bottom of its path, the ball strikes a 2.6e-kg steel block initially at rest on a frictionless surface. The collision is elastic. Find (a) the speed of the ball and (b) the speed of the block, both just after the collision.

```python
# Set up the figure with a dark background
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(12, 8))

# Surface
ax.plot([-2, 8], [0, 0], 'w-', lw=3, label='Frictionless Surface')

# Cord and swing path
theta = np.linspace(np.pi/2, 0, 100)
r = 0.687
x_cord = r * np.cos(theta)
y_cord = r * np.sin(theta) + 1
ax.plot(x_cord + 2, y_cord, 'w--', lw=2, label='Cord Path')
ax.plot(2, 1.687, 'o', ms=5, c='white')  # Pivot

# Ball (before collision)
ax.plot(2, 1, 'o', ms=10, c='cyan', label='Ball (0.514 kg)')
ax.arrow(2, 1, 0.5, 0, head_width=0.1, head_length=0.2, fc='cyan', ec='cyan')
ax.text(2.6, 1.2, 'v₁ = 3.67 m/s', color='cyan', fontsize=12)

# Block (before collision)
block = plt.Rectangle((3, 0), 1, 0.5, fc='gray', ec='white', label='Block (2.60 kg)')
ax.add_patch(block)

# Ball (after collision)
ax.plot(1.5, 1, 'o', ms=10, c='red')
ax.arrow(1.5, 1, -0.5, 0, head_width=0.1, head_length=0.2, fc='red', ec='red')
ax.text(1, 1.2, 'u₁ = -2.46 m/s', color='red', fontsize=12)

# Block (after collision)
block_after = plt.Rectangle((4.5, 0), 1, 0.5, fc='gray', ec='red')
ax.add_patch(block_after)
ax.arrow(4.5, 0.25, 0.5, 0, head_width=0.1, head_length=0.2, fc='red', ec='red')
ax.text(5, 0.45, 'u₂ = 1.21 m/s', color='red', fontsize=12)

# Styling
ax.set_xlim(-2, 8)
ax.set_ylim(-0.5, 2.5)
ax.set_aspect('equal')
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel('Horizontal', fontsize=14, color='white')
ax.set_ylabel('Vertical', fontsize=14, color='white')
ax.set_title('Elastic Collision: Ball and Block', fontsize=16, color='white')
ax.legend(fontsize=12, loc='upper left')
ax.grid(True, ls='--', alpha=0.3)

# Remove spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('white')

plt.show()
```

To push a 25 kg crate up a 27Â° incline, a worker exerts a force of 120 N, parallel to the incline. As the crate slides 3.6 m, how much work is done on the crate by: (a) the worker, (b) the force of gravity, and (c) the normal force due to the incline?

```python
# Set up the figure with a dark background
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(10, 6))

# Incline
theta = np.deg2rad(27)
length = 4  # Scaled for plot
x = [0, length * np.cos(theta)]
y = [0, length * np.sin(theta)]
ax.plot(x, y, 'w-', lw=4, label='Incline (27Â°)')

# Ground
ax.plot([-1, 4], [0, 0], 'w-', lw=2, label='Ground')

# Crate
crate_size = 0.5
crate_x = (length - crate_size) * np.cos(theta)
crate_y = (length - crate_size) * np.sin(theta)
crate = plt.Rectangle((crate_x - crate_size/2, crate_y), crate_size, crate_size/2,
                     angle=27, fc='gray', ec='white', label='Crate (25 kg)')
ax.add_patch(crate)

# Forces
# Workerâs force
ax.arrow(crate_x, crate_y + crate_size/4, 0.7 * np.cos(theta), 0.7 * np.sin(theta),
         head_width=0.1, head_length=0.1, fc='red', ec='red')
ax.text(crate_x + 0.8 * np.cos(theta), crate_y + 0.8 * np.sin(theta) + 0.1,
        'F = 120 N', color='red', fontsize=12)

# Normal force
ax.arrow(crate_x, crate_y + crate_size/4, -0.5 * np.sin(theta), 0.5 * np.cos(theta),
         head_width=0.1, head_length=0.1, fc='green', ec='green')
ax.text(crate_x - 0.6 * np.sin(theta), crate_y + 0.6 * np.cos(theta) + 0.1,
        'N = 218 N', color='green', fontsize=12)

# Gravity
ax.arrow(crate_x, crate_y + crate_size/4, 0, -0.7, head_width=0.1, head_length=0.1,
         fc='cyan', ec='cyan')
ax.text(crate_x, crate_y - 0.8, 'mg = 245 N', color='cyan', fontsize=12, ha='center')

# Displacement
ax.arrow(crate_x - 0.5 * np.cos(theta), crate_y - 0.5 * np.sin(theta),
         0.5 * np.cos(theta), 0.5 * np.sin(theta),
         head_width=0.1, head_length=0.1, fc='white', ec='white')
ax.text(crate_x, crate_y + 0.1, '3.6 m', color='white', fontsize=12, ha='center')

# Styling
ax.set_xlim(-1, 4)
ax.set_ylim(-1, 2.5)
ax.set_aspect('equal')
ax.set_xlabel('x (m)', fontsize=14, color='white')
ax.set_ylabel('y (m)', fontsize=14, color='white')
ax.set_title('Crate on Incline', fontsize=16, color='white')
ax.legend(fontsize=12, loc='upper left')
ax.grid(True, ls='--', alpha=0.3)

# Remove spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('white')
ax.spines['bottom'].set_color('white')

plt.show()
```

```python
# Set up the figure with a dark background
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(12, 6))

# Ground
ax.plot([-2, 8], [0, 0], 'w-', lw=3, label='Ground')

# Cart
cart = plt.Rectangle((2, 0), 2, 1, fc='gray', ec='white', label='Cart')
ax.add_patch(cart)

# Horse (simplified as a point)
ax.plot(1, 0.5, 'o', ms=15, c='cyan', label='Horse')

# Force vector
theta = np.deg2rad(27)
ax.arrow(1, 0.5, 0.5 * np.cos(theta), 0.5 * np.sin(theta),
         head_width=0.1, head_length=0.2, fc='cyan', ec='cyan')
ax.text(1.6, 0.8, 'F = 42 lb, 27Â°', color='cyan', fontsize=12)

# Motion
ax.arrow(2, 1.2, 2, 0, head_width=0.1, head_length=0.2, fc='white', ec='white')
ax.text(3, 1.4, 'v = 6.20 mi/hr\n12 min', color='white', fontsize=12)

# Work and power note
ax.text(5, 2, 'W = 245,000 ftÂ·lb\nP = 0.619 hp',
        color='white', fontsize=12, ha='center',
        bbox=dict(facecolor='black', alpha=0.5))

# Styling
ax.set_xlim(-2, 8)
ax.set_ylim(-0.5, 2.5)
ax.set_aspect('equal')
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel('Horizontal', fontsize=14, color='white')
ax.set_ylabel('Vertical', fontsize=14, color='white')
ax.set_title('Horse Pulling Cart', fontsize=16, color='white')
ax.legend(fontsize=12, loc='upper left')
ax.grid(True, ls='--', alpha=0.3)

# Remove spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('white')

plt.show()
```

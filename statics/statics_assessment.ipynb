Weights of 2, 4, 6 and 8 N are fastened to the corners of a 2 m x 2 m frame. If it weighs 10 N, where would the fulcrum (the support point) be placed to balanced the frame and the weights?

```python
import matplotlib.pyplot as plt
import numpy as np

# Set up the figure with a dark background
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(8, 8))

# Define the square frame
frame_x = [0, 2, 2, 0, 0]
frame_y = [0, 0, 2, 2, 0]
ax.plot(frame_x, frame_y, 'w-', lw=2, label='Frame (10 N at center)')

# Weights and their positions
weights = [2, 4, 6, 8, 10]
positions = [(0, 0), (2, 0), (2, 2), (0, 2), (1, 1)]
labels = ['2 N', '4 N', '6 N', '8 N', '10 N']

# Plot weights
for (x, y), w, label in zip(positions, weights, labels):
    ax.scatter(x, y, s=100, c='cyan', zorder=5)
    ax.annotate(label, (x, y), xytext=(5, 5), textcoords='offset points',
                fontsize=12, color='cyan')

# Fulcrum position
fulcrum = (1, 38/30)
ax.scatter(*fulcrum, s=150, c='red', marker='^', label='Fulcrum', zorder=10)
ax.annotate(f'Fulcrum\n({fulcrum[0]:.1f}, {fulcrum[1]:.4f})', fulcrum,
            xytext=(10, -15), textcoords='offset points', fontsize=12, color='red')

# Styling
ax.set_xlim(-0.5, 2.5)
ax.set_ylim(-0.5, 2.5)
ax.set_aspect('equal')
ax.set_xlabel('x (m)', fontsize=14, color='white')
ax.set_ylabel('y (m)', fontsize=14, color='white')
ax.set_title('Balanced Frame with Weights', fontsize=16, color='white')
ax.legend(fontsize=12, loc='upper right')
ax.grid(True, ls='--', alpha=0.3)

# Remove top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('white')
ax.spines['bottom'].set_color('white')

plt.show()
```

A ladder 12 m long leans against a vertical frictionless wall with its lower end 3 m from the wall. The ladder weighs 40 N. Find the magnitude and direction of the force exerted on the lower end of the ladder.

```python
# Set up the figure with a dark background
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(10, 8))

# Ladder
x = [0, 3]
y = [0, 3 * np.sqrt(15)]
ax.plot(x, y, 'w-', lw=3, label='Ladder (40 N)')

# Ground and wall
ax.plot([0, 3], [0, 0], 'w-', lw=2, label='Ground')
ax.plot([3, 3], [0, 12], 'w-', lw=2, label='Wall')

# Forces
# Weight at center
ax.arrow(1.5, 5.8095, 0, -2, head_width=0.1, head_length=0.3, fc='cyan', ec='cyan')
ax.text(1.6, 4, '40 N', color='cyan', fontsize=12)

# Normal force from wall
ax.arrow(3, 11.619, -1, 0, head_width=0.3, head_length=0.3, fc='yellow', ec='yellow')
ax.text(2, 11.8, 'N_w  5.16 N', color='yellow', fontsize=12)

# Forces at base
ax.arrow(0, 0, 0.8, 0, head_width=0.3, head_length=0.3, fc='red', ec='red')
ax.text(0.9, 0.2, 'F_f  5.16 N', color='red', fontsize=12)
ax.arrow(0, 0, 0, 3, head_width=0.1, head_length=0.3, fc='green', ec='green')
ax.text(0.2, 3, 'F_N = 40 N', color='green', fontsize=12)

# Net force
ax.arrow(0, 0, 0.8, 6, head_width=0.2, head_length=0.3, fc='magenta', ec='magenta')
ax.text(1, 6, 'F  40.33 N, 82.6', color='magenta', fontsize=12)

# Styling
ax.set_xlim(-1, 10)
ax.set_ylim(-1, 15)
ax.set_aspect('equal')
ax.set_xlabel('x (m)', fontsize=14, color='white')
ax.set_ylabel('y (m)', fontsize=14, color='white')
ax.set_title('Ladder in Equilibrium', fontsize=16, color='white')
ax.legend(fontsize=12, loc='best')
ax.grid(True, ls='--', alpha=0.3)

# Remove top/right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('white')
ax.spines['bottom'].set_color('white')

plt.show()
```

A uniform 25 N beam 6 m long is carried by two men A and B, one at each end of the beam.

```python
# Set up the figure with a dark background
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(10, 4))

# Beam
ax.plot([0, 6], [0, 0], 'w-', lw=4, label='Beam (25 N)')

# Forces
# As force at x = 0
ax.arrow(0, 0, 0, 0.5, head_width=0.1, head_length=0.2, fc='green', ec='green')
ax.text(0, 0.6, 'F_A = 20 N', color='green', fontsize=12, ha='center')

# Bs force at x = 6
ax.arrow(6, 0, 0, 0.8, head_width=0.1, head_length=0.2, fc='red', ec='red')
ax.text(6, 0.9, 'F_B = 50 N', color='red', fontsize=12, ha='center')

# Beam weight at x = 3
ax.arrow(3, 0, 0, -0.5, head_width=0.1, head_length=0.2, fc='cyan', ec='cyan')
ax.text(3, -0.6, '25 N', color='cyan', fontsize=12, ha='center')

# Load at x = 5
ax.arrow(5, 0, 0, -0.7, head_width=0.1, head_length=0.2, fc='yellow', ec='yellow')
ax.text(5, -0.8, '45 N', color='yellow', fontsize=12, ha='center')

# Labels for A and B
ax.text(0, -0.2, 'A', color='white', fontsize=14, ha='center')
ax.text(6, -0.2, 'B', color='white', fontsize=14, ha='center')

# Styling
ax.set_xlim(-1, 7)
ax.set_ylim(-1, 1.5)
ax.set_xticks(np.arange(0, 7, 1))
ax.set_yticks([])
ax.set_xlabel('Distance (m)', fontsize=14, color='white')
ax.set_title('Beam in Equilibrium', fontsize=16, color='white')
ax.legend(fontsize=12, loc='upper left')
ax.grid(True, ls='--', alpha=0.3)

# Remove spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('white')

plt.show()
```

A ladder, whose length is 10 m and whose mass is 40 kg rests against a frictionless vertical wall. Its upper end is a distance of 7.7 m above the ground. The center of mass of the ladder is 1/3 of the way up the ladder. The coefficient of static friction between the ladder and the ground is 0.53. If a carpenter climbs 85% of the way up the ladder before it starts to slip, find the mass of the carpenter.

```python
# Set up the figure with a dark background
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(10, 6))

# Ground (x-axis from -1 to 7.5 at y = 0)
ax.plot([-1, 7.5], [0, 0], 'w-', lw=2, label='Ground')

# Wall (vertical line at x = 6.38 from y = 0 to 8.5)
ax.plot([6.38, 6.38], [0, 8.5], 'w-', lw=2, label='Frictionless Wall')

# Ladder
ax.plot([0, 6.38], [0, 7.7], 'w-', lw=4, label='Ladder (392 N)')

# Forces
# Ground normal and friction at x = 0
ax.arrow(0, 0, 0, 1, head_width=0.2, head_length=0.3, fc='green', ec='green')
ax.text(0, 1.2, 'N = 965 N', color='green', fontsize=12, ha='center')
ax.arrow(0, 0, 1, 0, head_width=0.2, head_length=0.3, fc='red', ec='red')
ax.text(1, 0.2, 'F_f = 511 N', color='red', fontsize=12)

# Ladder weight at (2.12, 2.56)
ax.arrow(2.12, 2.56, 0, -0.8, head_width=0.2, head_length=0.3, fc='cyan', ec='cyan')
ax.text(2.12, 1.5, '392 N', color='cyan', fontsize=12, ha='center')

# Carpenter at (5.42, 6.55)
ax.arrow(5.42, 6.55, 0, -0.8, head_width=0.2, head_length=0.3, fc='yellow', ec='yellow')
ax.text(5.42, 5.5, '574 N', color='yellow', fontsize=12, ha='center')

# Wall normal at (6.38, 7.7)
ax.arrow(6.38, 7.7, -1, 0, head_width=0.2, head_length=0.3, fc='magenta', ec='magenta')
ax.text(5.5, 7.9, 'N_w = 511 N', color='magenta', fontsize=12)

# Labels
ax.text(5.42, 6.75, 'Carpenter', color='white', fontsize=12, ha='center')
ax.text(6.38, -0.3, 'Wall', color='white', fontsize=12, ha='center', rotation=90)
ax.text(-0.5, -0.3, 'Ground', color='white', fontsize=12, ha='center')

# Styling
ax.set_xlim(-1, 7.5)
ax.set_ylim(-1, 8.5)
ax.set_aspect('equal')
ax.set_xlabel('x (m)', fontsize=14, color='white')
ax.set_ylabel('y (m)', fontsize=14, color='white')
ax.set_title('Ladder at Slipping Point', fontsize=16, color='white')
ax.legend(fontsize=12, loc='upper left')
ax.grid(True, ls='--', alpha=0.3)

# Remove spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('white')
ax.spines['bottom'].set_color('white')

plt.show()
```
